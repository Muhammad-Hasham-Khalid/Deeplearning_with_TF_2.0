```
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```
(i_train,l_train),(i_test,l_test) = fashion_mnist.load_data()

#normalize
i_train = i_train / 255.0
i_test = i_test / 255.0

#making sequntial model and adding layers
model = tf.keras.Sequential(
    [
     tf.keras.layers.Flatten(), # making a single dimension vector from 2D
     tf.keras.layers.Dense(512, activation=tf.keras.activations.relu), #ouput of previous will be taken as input
     tf.keras.layers.Dense(10, activation=tf.keras.activations.softmax), #softmax is used for multiclassification
    ]
)

model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.sparse_categorical_crossentropy,
    metrics=['accuracy']
)

#training
model.fit(i_train, l_train, epochs=5, batch_size=128)

#testing
test_loss, test_acc = model.evaluate(i_test, l_test)
print("Testing accuracy is : ", test_acc)
```

```
plt.plot(model.predict(i_test)[0])
print(l_test) # Ankle boot
plt.show()
```
